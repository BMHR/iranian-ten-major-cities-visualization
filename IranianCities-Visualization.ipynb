# Visualization of Ten Major Iranian Cities Using Python

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = ["Tehran", "Mashhad", "Esfahan", "Karaj", "Shiraz", "Tabriz", "Ahvaz", "Qom", "Kerman", "Rasht"]

latitude = [51.2323, 59.3236, 51.4013, 50.5803, 52.3233, 46.1757, 48.4042, 50.5235, 57.0327, 49.3520]
congitude = [35.4120, 36.1935, 32.3955, 35.4945, 29.3636, 38.0454, 31.1817, 34.3824, 30.1545, 37.1628]
 
population = [8.693706, 3.001184, 1.961260, 1.592492, 1.565572, 1.558693, 1.184788, 1.201158, 0.537718, 0.679995] #millions
#Reported from 2016
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(llcrnrlon=40, llcrnrlat=25, urcrnrlon=65, urcrnrlat=40,
            resolution='l', projection='merc',
            lat_0=32.0, lon_0=53.0, lat_ts=20)

#   
m.drawcoastlines()
m.fillcontinents(color="lightgray", lake_color="lightblue")
m.drawparallels(np.arange(25, 40, 5), labels=[1, 1, 0, 1])
m.drawmeridians(np.arange(40, 65, 5), labels=[1, 1, 0, 1])


#    
color = ["deepskyblue", "gold", "lime", "violet", "blue", "cyan", "yellow", "red", "gray", "orange"]
for i, city in enumerate(cities):
    x, y = m(latitude[i], congitude[i])
    m.scatter(x, y, s=population[i] * 200, color=color[i], alpha=0.75)
    plt.text(x, y, city, fontsize=10, ha="center", color="black", rotation=10)
    
plt.title("Visualization of Ten Major Iranian Cities Using Python")
plt.show()
```
