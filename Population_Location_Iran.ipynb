# موقعیت جغرافیایی و جمعیت 10 شهر بزرگ ایران

```python
import matplotlib.pyplot as plt
```

```python
cities = ["Tehran", "Mashhad", "Esfahan", "Karaj", "Shiraz", "Tabriz", "Ahvaz", "Qom", "Kerman", "Rasht"]

latitude = [51.2323, 59.3236, 51.4013, 50.5803, 52.3233, 46.1757, 48.4042, 50.5235, 57.0327, 49.3520]
congitude = [35.4120, 36.1935, 32.3955, 35.4945, 29.3636, 38.0454, 31.1817, 34.3824, 30.1545, 37.1628]
 
population = [8.693706, 3.001184, 1.961260, 1.592492, 1.565572, 1.558693, 1.184788, 1.201158, 0.537718, 0.679995] #millions
#Reported from 2016
```

```python
color = ["deepskyblue", "gold", "lime", "violet", "blue", "cyan", "yellow", "red", "gray", "orange"]
for i in range(len(cities)):
    plt.scatter(latitude[i], congitude[i], color = color[i], label = cities[i], s = population[i]*1000, alpha = 0.6, edgecolors="black")

plt.title("The Population and Geographical Locations of 10 Major Iranian Cities")
plt.xlabel("Latitude")
plt.ylabel("Congitude")
plt.xlim([40 , 70])
plt.legend(markerscale = 0.4)
plt.show()
```
